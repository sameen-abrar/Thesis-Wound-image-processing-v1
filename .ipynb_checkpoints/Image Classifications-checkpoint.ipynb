```python
import cv2
import numpy as np

def create_wound_mask(image_path, lower_threshold, upper_threshold):
    # Read the image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to segment the wound region
    _, binary_mask = cv2.threshold(grayscale, lower_threshold, upper_threshold, cv2.THRESH_BINARY)

    # Invert the binary mask (wound region becomes white, background becomes black)
    inverted_mask = cv2.bitwise_not(binary_mask)

    return inverted_mask

# Example usage
image_path = 'datasets/Wound_dataset_1/Ingrown_nails/ingrown_nail (14).jpg'
lower_threshold = 100  # Adjust based on the characteristics of your images
upper_threshold = 255

wound_mask = create_wound_mask(image_path, lower_threshold, upper_threshold)

# Display the original image and the created mask
# cv2.imshow('Original Image', cv2.imread(image_path))
# cv2.imshow('Wound Mask', wound_mask)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
data = tf.keras.utils.image_dataset_from_directory(data_dir)
```

```python
data_iterator = data.as_numpy_iterator()
```

```python
batch = data_iterator.next()
```

```python
class_mapping = {
    0: 'Abrasions',
    1: 'Bruises',
    2: 'Burns',
    3: 'Cut',
    4: 'Ingrown_nails',
    5: 'Laceration',
    6: 'Stab_wound'
}

# batch[1] contains the numerical labels
numerical_labels = batch[1]

# Map numerical labels to class names
class_names = [class_mapping[label] for label in numerical_labels]

# Print the resulting class names
print(class_names)
```

```python
images_to_show = 5
fig, ax = plt.subplots(ncols=images_to_show, figsize=(20,20))
for idx, img in enumerate(batch[0][:images_to_show]):
    ax[idx].imshow(img.astype(int)) 
    # Set the title using the corresponding class name
    label_idx = batch[1][idx]
    class_name = class_mapping[label_idx]
    ax[idx].set_title(class_name)
```

```python
# Scale data to be between 0 and 1
scaled_data = data.map(lambda x, y: (x/255, y))
```

```python
scaled_data.as_numpy_iterator().next()[0].max()
```

```python
len(scaled_data)
```

```python
# define sizes for data
train_size = int(len(data)*.7)
val_size = int(len(data)*.2) + 1
test_size = int(len(data)*.1) + 1 # Used Post training
```

```python
train_size + val_size + test_size == len(scaled_data)
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
# end of preprocess
```

```python
len(test)
```

```python
# Build the deep learning model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
model = Sequential()
```

```python
# First layer needs input
# Relu converts any negative values to zero to make it linear
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dense(7, activation='softmax')) # For 7 classes of wounds.
```

```python
model.compile('adam', loss=tf.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
logdir = 'logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])
```

```python
hist.history
```

```python
# VIsualize losses
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
# Visualize accuracy
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
# Evaluation 
from tensorflow.keras.metrics import Precision, Recall, SparseCategoricalAccuracy
```

```python
precision = Precision()
recall = Recall()
accuracy = SparseCategoricalAccuracy()
```

```python
len(test)
```

```python
# Iterate through the test dataset
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)

    # Convert the predictions to class indices if needed
    yhat_classes = tf.argmax(yhat, axis=1)
                             
    precision.update_state(y, yhat_classes)
    recall.update_state(y, yhat_classes)
    accuracy.update_state(y, yhat)
```

```python
# Get the final evaluation results
precision_result = precision.result().numpy()
recall_result = recall.result().numpy()
accuracy_result = accuracy.result().numpy()

# Print or use the evaluation results as needed
print(f'Test Precision: {precision_result}')
print(f'Test Recall: {recall_result}')
print(f'Test Accuracy: {accuracy_result}')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

# Load the external image
external_image_path = 'datasets/External images for testing/abrasion.png'
external_image = cv2.imread(external_image_path)
external_image = cv2.resize(external_image, (256, 256))  # Resize to match the input size
external_image = external_image / 255.0  # Normalize to [0, 1]

# Expand dimensions to match the input shape expected by the model
external_image = np.expand_dims(external_image, axis=0)

# Make predictions
predictions = model.predict(external_image)

# Get the top N predictions
top_predictions = np.argsort(predictions[0])[::-1][:3]  # Adjust N as needed

# Map the class indices to the corresponding class labels
class_labels = ["Abrasions", "Bruises", "Burns", "Cut", "Ingrown_nails", "Laceration", "Stab_wound"]

# Print the top predicted classes and their probabilities
for i in top_predictions:
    predicted_class_label = class_labels[i]
    probability = predictions[0][i]
    print(f"Class: {predicted_class_label}, Probability: {probability}")

# Visualize the external image
plt.imshow(cv2.cvtColor(cv2.imread(external_image_path), cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = 'datasets/External images for testing/abrasion.png'
img = cv2.imread(image_path)

# Check if the image is loaded successfully
if img is None:
    raise ValueError(f"Error loading image at path: {image_path}")

# Resize the image
resize = tf.image.resize(img, (256, 256))

# Display the original and resized images
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(resize.numpy().astype(int))
plt.title('Resized Image')

plt.show()

# Normalize the resized image
resize = resize / 255.0

# Make predictions
yhat = model.predict(np.expand_dims(resize, 0))

# Map the class indices to the corresponding class labels
class_labels = ['Abrasion', 'Bruise', 'Burn', 'Cut', 'Ingrown_nail', 'Laceration', 'Stab_wound']

# Display the predicted probabilities for each class
for i, label in enumerate(class_labels):
    print(f'Probability for class "{label}": {yhat[0][i]}')

# Get the predicted class index
predicted_class_index = np.argmax(yhat[0])

# Print the predicted class and its probability
predicted_class_label = class_labels[predicted_class_index]
probability = yhat[0][predicted_class_index]
print(f'\nPredicted class: {predicted_class_label}, Probability: {probability}')

# Make a decision based on a threshold (e.g., 0.5)
if probability > 0.5:
    print(f'Predicted class is a {predicted_class_label}')
else:
    print(f'Predicted class is not a {predicted_class_label}')
```
